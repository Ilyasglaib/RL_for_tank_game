```python
def plot_rewards(rewards_given, episode, window_size=10):
    episodes = range(episode + 1)
    rewards_moving_avg = np.convolve(rewards_given[:episode + 1], np.ones(window_size)/window_size, mode='valid')
    plt.plot(episodes[window_size-1:], rewards_moving_avg)
    plt.xlabel('Episode')
    plt.ylabel('Moving Average Rewards')
    plt.title(f'Moving Average Rewards per Episode (Window Size = {window_size})')
    plt.grid(True)
    plt.show()

def plot_kill_streak(kill_streak_count, episode, window_size=10):
    episodes = range(episode + 1)
    rewards_moving_avg = np.convolve(kill_streak_count[:episode + 1], np.ones(window_size)/window_size, mode='valid')
    plt.plot(episodes[window_size-1:], rewards_moving_avg)
    plt.xlabel('Episode')
    plt.ylabel('Moving Average kill_streak')
    plt.title(f'Moving Average kill_streak per Episode (Window Size = {window_size})')
    plt.grid(True)
    plt.show()


def simulate_key_press(key):
    pygame.event.post(pygame.event.Event(KEYDOWN, {'key': key}))

def perform_action(action):
    if action==0:
        simulate_key_press(K_UP)
    if action==1:
        simulate_key_press(K_RIGHT)
    if action==2:
        simulate_key_press(K_DOWN)
    if action==3:
        simulate_key_press(K_LEFT)
    if action==5:
        simulate_key_press(K_SPACE)
```

```python
#hyperparameters
total_episodes = 2000
visualisation_frequency=100
batch_size=32
target_update = 20
memory = PrioritizedReplayBuffer(10000)
save_interval = 200  # Every 50 episodes
save_path = "checkpoints/"
```

```python
# Initialisation agent
input_shape = (12, 22, 22)  # pour 4 images RGB 22x22
num_actions = 6
agent = DQNAgent(input_shape=input_shape, num_actions=num_actions)
agent.model.load_state_dict(torch.load("C:/Users/DELL/Documents/Reinfrocement learning project/notebooks/checkpoints/model_ep900.pth")
)
rewards_given=np.zeros(total_episodes+1)
kill_streak_count=np.zeros(total_episodes+1)
```

```python
# Initialize pygame and the environment
pygame.init()


for episode in range (total_episodes+1):
    frame_stack = deque(maxlen=4)

    env = TankEnv()
    env.reset()

    # Set up the display
    screen_size = (env.max_x * 20, env.max_y * 20)  # Scale up the game screen
    screen = pygame.display.set_mode(screen_size)
    clock = pygame.time.Clock()


    current_frame=env.render() #The state now is 2 consecutive frames !
    for _ in range(4):
        frame_stack.append(current_frame)
    frame_state = np.concatenate(list(frame_stack), axis=2)

    done=False

    total_reward = 0
    

    while not done:
        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                done = True
        # Choose an action with epsilon greedy
        action = agent.select_action(frame_state)

        #perform action
        perform_action(action)

        # Get the reward
        next_state, reward,kill_streak, done,_ = env.step(action)

        #Get the next frame state
        next_frame = env.render()
        frame_stack.append(next_frame)
        next_frame_state = np.concatenate(list(frame_stack), axis=2)

        #render the game
        frame=next_frame.copy()
        frame = np.repeat(np.repeat(frame, 18, axis=0), 18, axis=1)  # Scale up the frame for visibility
        surface = pygame.surfarray.make_surface(frame)
        screen.blit(surface, (0, 0))
        pygame.display.flip()

        # Cap the frame rate
        clock.tick(9)

        # Stock the transition
        memory.push(frame_state, action, reward, next_frame_state, done)

        # Train model on batch
        agent.train_step_PER(memory, batch_size)

        frame_state=next_frame_state
        total_reward += reward

    # Update the target model
    if episode % target_update == 0:
        agent.update_target()

    #Display rewards and kill_streaks
    rewards_given[episode]=total_reward
    if kill_streak>=20:
            print("The game has been cleared")
    kill_streak_count[episode]=kill_streak
    if episode%visualisation_frequency==0 and episode>10:
        plot_rewards(rewards_given,episode,window_size=10)
        plot_kill_streak(kill_streak_count,episode,window_size=10)

    #save model every save_interval episode
    os.makedirs(save_path, exist_ok=True)
    if episode % save_interval == 0:
        torch.save(agent.model.state_dict(), os.path.join(save_path, f"model_ep{episode}.pth"))
        print(f"[INFO] Model has been saved at episode {episode}")

pygame.quit()
```

```python
#Evaluation mode i.e without exploration or training

agent.model.eval()  # Put the network on evaluation mode (no dropout, batchnorm, etc.)

env = TankEnv()
env.reset()
frame_stack = deque(maxlen=4)

# Set up the display
screen_size = (env.max_x * 20, env.max_y * 20)  # Scale up the game screen
screen = pygame.display.set_mode(screen_size)
clock = pygame.time.Clock()

current_frame=env.render() #The state now is the frame !

for _ in range(4):
    frame_stack.append(current_frame)
frame_state = np.concatenate(list(frame_stack), axis=2)

done=False
total_reward=0

while not done:
    state_tensor = torch.tensor(frame_state, dtype=torch.float32).permute(2, 0, 1).unsqueeze(0)
    with torch.no_grad():
        q_values = agent.model(state_tensor)
        action = torch.argmax(q_values).item()
        
    #perform action
    perform_action(action)

    # Get the reward
    next_state, reward,kill_streak, done,_ = env.step(action)

    next_frame = env.render()
    frame_stack.append(next_frame)
    next_frame_state = np.concatenate(list(frame_stack), axis=2)

    #render the game
    frame=next_frame.copy()
    frame = np.repeat(np.repeat(frame, 18, axis=0), 18, axis=1)  # Scale up the frame for visibility
    surface = pygame.surfarray.make_surface(frame)
    screen.blit(surface, (0, 0))
    pygame.display.flip()

    # Cap the frame rate
    clock.tick(9)
    
    frame_state=next_frame_state
    total_reward += reward

print(f"[EVAL] Total reward : {total_reward}")
print(f"[EVAL] Final kill streak : {kill_streak}")
```
